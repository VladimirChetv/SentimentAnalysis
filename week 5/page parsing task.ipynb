## Парсинг веб-страниц

В этом задании вы потренеруетесь парсить веб-страницы. Это умение пригодится вам, когда возникнет потребность самостоятельно собрать выборку для построения модели, будь то тексты или какая-то другая информация с общедоступных веб-страниц. В рамках данного проекта это случится уже в задании следующей недели.

```python
import requests
import bs4
```

1. Из статьи https://en.wikipedia.org/wiki/Bias-variance_tradeoff все заголовки верхнего уровня


```python
req = requests.get('https://en.wikipedia.org/wiki/Bias-variance_tradeoff')
req
```

```python
req.text
```

```python
parser = bs4.BeautifulSoup(req.text, 'lxml')
parser
```

Найдем заголовок верхнего уровня, т.е. h1

```python
y = parser.findAll('h1')
for ans in y:
    print (ans.text)
```

2. Со страницы https://en.wikipedia.org/wiki/Category:Machine_learning_algorithms названия всех статей в категории Machine Learning Algorithms

```python
req = requests.get('https://en.wikipedia.org/wiki/Category:Machine_learning_algorithms')
req.text
```

```python
parser = bs4.BeautifulSoup(req.text, 'lxml')
parser
```

Парсим страницу и находим названия всех статей, выводим также номер статьи в списке. 

```python
y = parser.findAll('div', attrs={'class':'mw-category-group'})
counter = 0
for res in y:
    y1 = res.findAll('a')
    for res1 in y1:
        counter += 1
        print (counter, res1.attrs['title'])
```
